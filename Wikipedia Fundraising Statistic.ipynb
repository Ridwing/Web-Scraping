```python
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/Wikipedia:Fundraising_statistics"

tables = pd.read_html(url)
```

```python
type(tables)
```

```python
len(tables)
```

```python
fundraising = tables[0]
```

```python
type(fundraising)
```

```python
fundraising.head()
```

```python
# check data type

fundraising.dtypes
```

```python
# turning dataset into CSV

fundraising['rev'] = fundraising['Revenue'].str[1:]            #removing $
fundraising['rev'] = fundraising['rev'].str.replace(',','')    #removing comma
fundraising
```

```python
# as datatypes are object/str, manipulate data being in the same type, after manipulating convert it into numeric

fundraising['exp'] = fundraising['Expenses'].str[1:]
fundraising['exp'] = fundraising['exp'].str.replace(',','')
fundraising['asset_rise'] = fundraising['Asset rise'].str[1:]
fundraising['asset_rise'] = fundraising['asset_rise'].str.replace(',','')
fundraising['total_asset'] = fundraising['Net assets atend of year'].str[1:]
fundraising['total_asset'] = fundraising['total_asset'].str.replace(',','')

fundraising
```

```python
# changing datatypes from object to integer

fundraising['rev'] = pd.to_numeric(fundraising['rev'])
fundraising['exp'] = pd.to_numeric(fundraising['exp'])
fundraising['asset_rise'] = pd.to_numeric(fundraising['asset_rise'])
fundraising['total_asset'] = pd.to_numeric(fundraising['total_asset'])

fundraising
```

```python
fundraising.dtypes
```

```python
# delete columns by name with .drop()

fundraising = fundraising.drop(["Expenses", "Asset rise", "Net assets atend of year"],axis = 1)            

fundraising
```
